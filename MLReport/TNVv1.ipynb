```python
# Importing necessary libraries
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function to visualize the multidimensional thought network of Alafia
def plot_thought_network():
    """
    Creates and plots a directed graph representing Alafia's thought network.
    The graph includes categories of thoughts and their sub-components.
    """
    
    # Create a directed graph
    G = nx.DiGraph()

    # Adding nodes and their categories (directions)
    G.add_node("Future Aspirations")
    G.add_node("Emotional Detachment")
    G.add_node("Control Over Emotions")
    G.add_node("Avoidance of Conflict")

    # Adding sub-nodes and their connections for each direction

    # Future Aspirations Branch
    G.add_node("Career & Future Husband")
    G.add_edge("Future Aspirations", "Career & Future Husband")

    # Emotional Detachment Branch
    G.add_node("Coldness & Emotional Walls")
    G.add_node("Distance from Anindya")
    G.add_node("Indifference Toward Anindya")
    G.add_edge("Emotional Detachment", "Coldness & Emotional Walls")
    G.add_edge("Coldness & Emotional Walls", "Distance from Anindya")
    G.add_edge("Distance from Anindya", "Indifference Toward Anindya")

    # Control Over Emotions Branch
    G.add_node("Selective Emotional Access")
    G.add_node("Compartmentalization")
    G.add_node("Rationalizing Distance")
    G.add_edge("Control Over Emotions", "Selective Emotional Access")
    G.add_edge("Selective Emotional Access", "Compartmentalization")
    G.add_edge("Compartmentalization", "Rationalizing Distance")

    # Avoidance of Conflict Branch
    G.add_node("Evasion & Deflection")
    G.add_node("Deliberate Ambiguity")
    G.add_node("Playfulness or Manipulation")
    G.add_edge("Avoidance of Conflict", "Evasion & Deflection")
    G.add_edge("Evasion & Deflection", "Deliberate Ambiguity")
    G.add_edge("Deliberate Ambiguity", "Playfulness or Manipulation")

    # Manually setting positions for consistent layout
    pos = {
        "Future Aspirations": (0, 1),
        "Career & Future Husband": (0, 0.5),
        "Emotional Detachment": (1, 1),
        "Coldness & Emotional Walls": (1, 0.75),
        "Distance from Anindya": (1, 0.5),
        "Indifference Toward Anindya": (1, 0.25),
        "Control Over Emotions": (2, 1),
        "Selective Emotional Access": (2, 0.75),
        "Compartmentalization": (2, 0.5),
        "Rationalizing Distance": (2, 0.25),
        "Avoidance of Conflict": (3, 1),
        "Evasion & Deflection": (3, 0.75),
        "Deliberate Ambiguity": (3, 0.5),
        "Playfulness or Manipulation": (3, 0.25)
    }

    plt.figure(figsize=(12, 8))

    # Draw nodes with labels
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", arrows=True, arrowstyle='->', arrowsize=15)
    plt.title("Thought Network Visualization for Alafia", size=15)
    plt.savefig("thought_network.png")
    plt.show()
```

```python
# Function to plot Alafia's reactions on a graph with x and y axis
def plot_reactions():
    # X axis is a scale for external factors, Y axis is her reactions
    external_factors = np.linspace(0, 10, 100)
    
    # Define reactions under conditions
    reactions = 0.2 * external_factors**2 - 1.5 * external_factors + 5  # An arbitrary reaction model
    
    # Plotting the graph
    plt.figure(figsize=(8, 6))
    plt.plot(external_factors, reactions, label='Alafiaâ€™s Reactions', color='blue')
    plt.axhline(y=5, color='gray', linestyle='--', label='Baseline (Neutral Reaction)')
    plt.axvline(x=7, color='red', linestyle='--', label='Critical Point (High External Pressure)')
    
    # Labeling the graph
    plt.xlabel("External Factors (Pressure, Emotional Confrontation, Career Influence)", fontsize=12)
    plt.ylabel("Reaction Intensity", fontsize=12)
    plt.title("Graph of Alafia's Reactions under Different Parameters", fontsize=14)
    plt.legend()
    plt.show()
```

```python
def plot_trust_probability():
    # Probability conditions
    p_initial_mistrust = 0.8  # Initial mistrust (80% chance)
    p_self_preservation_barrier = 0.7  # Self-preservation causing emotional distance
    p_earn_trust_action = 0.4  # Chance of Anindya doing something to gain trust
    p_time_effect = 0.6  # Effect of time in healing
    p_communication = 0.5  # Open communication as a factor
    
    # Total probability formula
    p_gain_trust = (1 - p_initial_mistrust) * p_earn_trust_action * p_time_effect * p_communication
    
    # Plotting the trust possibility graph
    conditions = ['Initial Mistrust', 'Self-Preservation Barrier', 'Earn Trust Action', 'Time Effect', 'Open Communication']
    probabilities = [p_initial_mistrust, p_self_preservation_barrier, p_earn_trust_action, p_time_effect, p_communication]
    
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plot probability as a bar chart
    ax.bar(conditions, probabilities, color='blue')
    ax.set_ylim([0, 1])
    ax.set_title("Probabilities Affecting Anindya's Chances of Gaining Alafia's Trust", fontsize=14)
    ax.set_ylabel("Probability")
    
    # Rotate x-axis labels to 90 degrees
    plt.xticks(rotation=90)
    
    # Display total possibility result
    total_trust_chance = f"Total Trust Chance: {p_gain_trust:.2f} (i.e., {p_gain_trust * 100:.1f}%)"
    plt.text(0.5, 0.9, total_trust_chance, fontsize=12, ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()  # Adjust layout for rotated labels
    plt.show()
```

```python
# Call the functions
plot_thought_network()  # Visualize the multidimensional thought network
plot_reactions()        # Plot reactions under parameters
plot_trust_probability()# Plot the probability of regaining trust
```
